```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup as bs
import requests
import csv

import time
import random

from fake_useragent import UserAgent
```

```python
def write_csv(data):
    with open('avito_room.csv', 'a', encoding='utf-8') as f:
        writer = csv.writer(f)
        writer.writerow([data['title'], data['url'], data['price'], data['address'],
                        data['dist_metro'],data['agent'], data['discribe_list'], data['metro']])
```

```python
def get_html(url):
    
    time.sleep(random.randint(8, 15)) 

    r = requests.get(url, headers={'User-Agent': UserAgent().chrome})
    print(r.status_code)
    print('0')
    return r.text
```
